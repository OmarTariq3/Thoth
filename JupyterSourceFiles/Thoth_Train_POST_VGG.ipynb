#### 2_b) Undersampling over represented classes
1. **Identify High Percentage Classes**: Determine which classes have the highest number of samples.
2. **Calculate Undersampling Threshold**: Decide on a threshold for the maximum number of samples per class.
3. **Randomly Select Samples**: For classes above the threshold, randomly select a subset of samples to reduce the count.
4. **Create a Balanced Dataset**: Combine the undersampled classes with the minority classes to form a new, balanced dataset.

```python
import os
import random
import shutil
from collections import Counter

# Path to the all_data directory
all_data_path = r"C:\Root\علم و عمل\Graduation project\my_models\Thoth\Dataset\all_data"

# Get the list of all image files
image_files = [f for f in os.listdir(all_data_path) if os.path.isfile(os.path.join(all_data_path, f))]

# Extract class labels from filenames (assuming class labels are part of the filenames)
class_labels = [f.split('_')[1].replace('.png', '') for f in image_files]

# Count the occurrences of each class
class_counts = Counter(class_labels)

# Calculate the mean frequency of the class distributions
mean_frequency = sum(class_counts.values()) / len(class_counts)

# Identify classes that have a frequency higher than the mean
classes_to_undersample = {label: count for label, count in class_counts.items() if count > mean_frequency}

# Create a new directory for undersampled data
undersampled_data_path = r"C:\Root\علم و عمل\Graduation project\my_models\Thoth\Dataset\undersampled_data2"
os.makedirs(undersampled_data_path, exist_ok=True)

# Copy images, undersampling the high-frequency classes
for label in class_counts:
    class_images = [f for f in image_files if label == f.split('_')[1].replace('.png', '')]

    if label in classes_to_undersample:
        # Randomly select a subset of images to match the mean frequency
        class_images = random.sample(class_images, int(mean_frequency))

    for img in class_images:
        src_path = os.path.join(all_data_path, img)
        dest_path = os.path.join(undersampled_data_path, img)
        shutil.copy(src_path, dest_path)

print(f"Undersampled data created at {undersampled_data_path}")

# Verify the new distribution
new_image_files = [f for f in os.listdir(undersampled_data_path) if os.path.isfile(os.path.join(undersampled_data_path, f))]
new_class_labels = [f.split('_')[1].replace('.png', '') for f in new_image_files]
new_class_counts = Counter(new_class_labels)

# Print the new class distribution
print("New Class Distribution (Descending Order):")
sorted_new_class_counts = sorted(new_class_counts.items(), key=lambda item: item[1], reverse=True)
for label, count in sorted_new_class_counts:
    percentage = (count / sum(new_class_counts.values())) * 100
    print(f"{label}: {count} ({percentage:.2f}%)")



```

#### 2_d) Data Augmentation of underrepresented classes (Filpping, Rotation, and Mirroring)

```python
import os
import random
import shutil
from collections import Counter
from PIL import Image, ImageOps

# Path to the all_data directory
all_data_path = r"C:\Root\علم و عمل\Graduation project\my_models\Thoth\Dataset\all_data"

# Get the list of all image files
image_files = [f for f in os.listdir(all_data_path) if os.path.isfile(os.path.join(all_data_path, f))]

# Extract class labels from filenames (assuming class labels are part of the filenames)
class_labels = [f.split('_')[1].replace('.png', '') for f in image_files]

# Count the occurrences of each class
class_counts = Counter(class_labels)

# Calculate the mean frequency of the class distributions
mean_frequency = sum(class_counts.values()) / len(class_counts)

# Identify classes that have a frequency higher than the mean
classes_to_undersample = {label: count for label, count in class_counts.items() if count > mean_frequency}

# Identify classes that have a frequency lower than the mean
classes_to_augment = {label: count for label, count in class_counts.items() if count < mean_frequency}

# Create a new directory for undersampled data
undersampled_data_path = r"C:\Root\علم و عمل\Graduation project\my_models\Thoth\Dataset\undersampled_data2"
os.makedirs(undersampled_data_path, exist_ok=True)

# Copy images, undersampling the high-frequency classes
for label in class_counts:
    class_images = [f for f in image_files if label == f.split('_')[1].replace('.png', '')]

    if label in classes_to_undersample:
        # Randomly select a subset of images to match the mean frequency
        class_images = random.sample(class_images, int(mean_frequency))

    for img in class_images:
        src_path = os.path.join(all_data_path, img)
        dest_path = os.path.join(undersampled_data_path, img)
        shutil.copy(src_path, dest_path)

# Data augmentation for classes under the mean threshold
def augment_image(image_path, save_path, augment_count):
    image = Image.open(image_path)
    augmentations = [
        ImageOps.mirror,
        ImageOps.flip,
        lambda img: img.rotate(90),
        lambda img: img.rotate(180),
        lambda img: img.rotate(270)
    ]

    for i in range(augment_count):
        aug_img = augmentations[i % len(augmentations)](image)
        aug_img.save(os.path.join(save_path, f"{os.path.splitext(os.path.basename(image_path))[0]}_aug{i}.png"))

# Augment images for underrepresented classes
for label, count in classes_to_augment.items():
    class_images = [f for f in image_files if label == f.split('_')[1].replace('.png', '')]
    augment_count = int(mean_frequency) - count

    for img in class_images:
        src_path = os.path.join(all_data_path, img)
        dest_path = os.path.join(undersampled_data_path, img)
        shutil.copy(src_path, dest_path)

        # Apply augmentations
        augment_image(src_path, undersampled_data_path, augment_count)

print(f"Augmented data added to {undersampled_data_path}")

# Verify the new distribution
new_image_files = [f for f in os.listdir(undersampled_data_path) if os.path.isfile(os.path.join(undersampled_data_path, f))]
new_class_labels = [f.split('_')[1].replace('.png', '') for f in new_image_files]
new_class_counts = Counter(new_class_labels)

# Print the new class distribution
print("New Class Distribution (Descending Order):")
sorted_new_class_counts = sorted(new_class_counts.items(), key=lambda item: item[1], reverse=True)
for label, count in sorted_new_class_counts:
    percentage = (count / sum(new_class_counts.values())) * 100
    print(f"{label}: {count} ({percentage:.2f}%)")
```

#### 2_f) Using Stratified Sampling to make the Train, Val, and Test sets.
- Why Stratified Sampling?
    - Preserves Class Distribution: Ensures that each subset (train, validation, test) has a similar distribution of classes as the original dataset.
    - Improved Model Evaluation: Provides a more accurate evaluation of the model’s performance across all classes.
    - Prevents Bias: Prevents the model from being biased towards the majority class in any of the subsets.

```python
import os
import shutil
from sklearn.model_selection import train_test_split
from collections import Counter

# Path to the all_data directory
all_data_path = r"C:\Root\علم و عمل\Graduation project\my_models\Thoth\Dataset\undersampled_data2"

# Get the list of all image files
image_files = [f for f in os.listdir(all_data_path) if os.path.isfile(os.path.join(all_data_path, f))]

# Extract class labels from filenames (assuming class labels are part of the filenames)
class_labels = [f.split('_')[1].replace('.png', '') for f in image_files]

# Split data into train, validation, and test sets using stratified sampling
train_files, test_files, train_labels, test_labels = train_test_split(
    image_files, class_labels, test_size=0.2, stratify=class_labels, random_state=42)

train_files, val_files, train_labels, val_labels = train_test_split(
    train_files, train_labels, test_size=0.25, stratify=train_labels, random_state=42)  # 0.25 x 0.8 = 0.2

# Create directories for train, val, and test sets
train_path = r"C:\Root\علم و عمل\Graduation project\my_models\Thoth\Dataset\train"
val_path = r"C:\Root\علم و عمل\Graduation project\my_models\Thoth\Dataset\val"
test_path = r"C:\Root\علم و عمل\Graduation project\my_models\Thoth\Dataset\test"

os.makedirs(train_path, exist_ok=True)
os.makedirs(val_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)

# Function to copy files to their respective directories
def copy_files(file_list, dest_path):
    for file in file_list:
        src_path = os.path.join(all_data_path, file)
        dest_file_path = os.path.join(dest_path, file)
        os.makedirs(os.path.dirname(dest_file_path), exist_ok=True)
        shutil.copy(src_path, dest_file_path)

# Copy files to respective directories
copy_files(train_files, train_path)
copy_files(val_files, val_path)
copy_files(test_files, test_path)

print(f"Files copied to train, val, and test directories.")
```

#### 3_c) Run YOLOv8 Generate Bounding Boxes in YOLO format





```python
import os
from pathlib import Path
from ultralytics import YOLO
import cv2
import numpy as np

# Load the YOLOv8 model
model = YOLO('yolov8s.pt')  # You can choose other versions like YOLOv8n.pt, YOLOv8m.pt, YOLOv8l.pt, YOLOv8x.pt

# Define paths to your dataset
train_path = r"/content/Thoth/Dataset/train"
val_path = r"/content/Thoth/Dataset/val"
test_path = r"/content/Thoth/Dataset/test"

# Directories to save predictions
pred_train_dir = Path('/content/Thoth/predictions/yolov8/train')
pred_val_dir = Path('/content/Thoth/predictions/yolov8/val')
pred_test_dir = Path('/content/Thoth/predictions/yolov8/test')

pred_train_dir.mkdir(parents=True, exist_ok=True)
pred_val_dir.mkdir(parents=True, exist_ok=True)
pred_test_dir.mkdir(parents=True, exist_ok=True)

def draw_bounding_boxes(image, boxes, labels):
    for box in boxes:
        class_id = int(box.cls)
        x_center, y_center, width, height = box.xywhn[0]
        x_center, y_center, width, height = x_center * image.shape[1], y_center * image.shape[0], width * image.shape[1], height * image.shape[0]
        x_min = int(x_center - width / 2)
        y_min = int(y_center - height / 2)
        x_max = int(x_center + width / 2)
        y_max = int(y_center + height / 2)

        color = (0, 255, 0)  # Green color for bounding box
        thickness = 2
        cv2.rectangle(image, (x_min, y_min), (x_max, y_max), color, thickness)
        label = f"{labels[class_id]}"
        cv2.putText(image, label, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
    return image

# Function to generate predictions using YOLOv8
def run_inference(source_dir, output_dir):
    image_paths = list(Path(source_dir).rglob('*.png')) + list(Path(source_dir).rglob('*.jpg'))
    for img_path in image_paths:
        results = model(img_path)
        image = cv2.imread(str(img_path))
        for result in results:
            if result.boxes:
                image_with_boxes = draw_bounding_boxes(image, result.boxes, result.names)
                result_path = output_dir / (img_path.stem + "_result.jpg")
                cv2.imwrite(str(result_path), image_with_boxes)

            # Save bounding box information in a .txt file in YOLO format
            txt_path = output_dir / (img_path.stem + '.txt')
            with open(txt_path, 'w') as f:
                if result.boxes:
                    for box in result.boxes:
                        class_id = int(box.cls)
                        x_center, y_center, width, height = box.xywhn[0]
                        f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

# Run inference on train, val, and test sets
run_inference(train_path, pred_train_dir)
run_inference(val_path, pred_val_dir)
run_inference(test_path, pred_test_dir)

print("Bounding boxes generated for train, val, and test sets.")


```

#### 3_d) Convert YOLO to COCO Format

```python
import json
import shutil
from collections import defaultdict
from PIL import Image

# Directories for COCO format
coco_train_dir = Path('/content/Thoth/predictions/coco/train')
coco_val_dir = Path('/content/Thoth/predictions/coco/val')
coco_test_dir = Path('/content/Thoth/predictions/coco/test')

coco_train_dir.mkdir(parents=True, exist_ok=True)
coco_val_dir.mkdir(parents=True, exist_ok=True)
coco_test_dir.mkdir(parents=True, exist_ok=True)

# Function to convert YOLO format to COCO format
def yolo_to_coco(yolo_dir, coco_dir, images, annotations, image_id, annotation_id):
    for txt_file in yolo_dir.rglob('*.txt'):
        img_file = txt_file.with_suffix('.png')  # or .jpg if that's the case
        if not img_file.exists():
            continue

        shutil.copy(img_file, coco_dir / img_file.name)

        with open(txt_file, 'r') as f:
            img = Image.open(img_file)
            img_width, img_height = img.size
            for line in f:
                parts = line.strip().split()
                if len(parts) != 5:
                    continue  # Skip invalid entries
                class_id, x_center, y_center, width, height = map(float, parts)
                x_center *= img_width
                y_center *= img_height
                width *= img_width
                height *= img_height
                x_min = x_center - width / 2
                y_min = y_center - height / 2

                images.append({
                    "id": image_id,
                    "file_name": img_file.name,
                    "width": img_width,
                    "height": img_height
                })

                annotations.append({
                    "id": annotation_id,
                    "image_id": image_id,
                    "category_id": int(class_id),
                    "bbox": [x_min, y_min, width, height],
                    "area": width * height,
                    "iscrowd": 0
                })

                annotation_id += 1
        image_id += 1

    return images, annotations, image_id, annotation_id

# Initialize COCO format data
coco_data = {
    "images": [],
    "annotations": [],
    "categories": [{"id": i, "name": f"class_{i}"} for i in range(80)]  # Assuming 80 classes
}

# Convert and save COCO format data
image_id, annotation_id = 1, 1
coco_data['images'], coco_data['annotations'], image_id, annotation_id = yolo_to_coco(pred_train_dir, coco_train_dir, coco_data['images'], coco_data['annotations'], image_id, annotation_id)
coco_data['images'], coco_data['annotations'], image_id, annotation_id = yolo_to_coco(pred_val_dir, coco_val_dir, coco_data['images'], coco_data['annotations'], image_id, annotation_id)
coco_data['images'], coco_data['annotations'], image_id, annotation_id = yolo_to_coco(pred_test_dir, coco_test_dir, coco_data['images'], coco_data['annotations'], image_id, annotation_id)

# Save COCO annotations
with open('/content/Thoth/predictions/coco/annotations.json', 'w') as f:
    json.dump(coco_data, f)

print("Bounding boxes saved in COCO format for SSD and Faster R-CNN.")
```

#### 5_b) Import Libraries, Setup Dataset, and Creat Dataloaders for

```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision.io import read_image
from torchvision.transforms import ToTensor

class CustomDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        self.image_dir = image_dir
        self.image_files = [f for f in os.listdir(image_dir) if os.path.isfile(os.path.join(image_dir, f))]
        self.transform = transform

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_name = self.image_files[idx]
        img_path = os.path.join(self.image_dir, img_name)
        image = read_image(img_path)

        # Extract label from the filename
        label = img_name.split('_')[1].split('.')[0]  # Assumes format "unique_number_classlabel.png"
        label = int(label)  # Convert label to integer

        if self.transform:
            image = self.transform(image)

        return image, label

def create_dataloader(images_dir, batch_size=16, shuffle=True):
    dataset = CustomDataset(images_dir, transform=ToTensor())
    return DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)

# Define dataset paths
train_images_dir = '/content/Thoth/Dataset/train/images'
val_images_dir = '/content/Thoth/Dataset/val/images'
test_images_dir = '/content/Thoth/Dataset/test/images'
data_yaml_path = '/content/Thoth/Dataset/data.yaml'

# Create DataLoaders
train_loader = create_dataloader(train_images_dir)
val_loader = create_dataloader(val_images_dir, shuffle=False)

```

#### 6_b) Define the Custom Dataset

```python
# Define the number of classes (e.g., 171 classes + 1 background class)
num_classes = 172

# Define custom dataset class
class CustomDataset(Dataset):
    def __init__(self, images_dir, labels_dir, transforms=None):
        self.images_dir = images_dir
        self.labels_dir = labels_dir
        self.transforms = transforms
        self.imgs = list(sorted(os.listdir(images_dir)))
        self.labels = list(sorted(os.listdir(labels_dir)))

    def __getitem__(self, idx):
        img_path = os.path.join(self.images_dir, self.imgs[idx])
        label_path = os.path.join(self.labels_dir, self.labels[idx])
        img = Image.open(img_path).convert("RGB")

        with open(label_path) as f:
            annotation = json.load(f)

        boxes = torch.tensor([obj["bbox"] for obj in annotation["annotations"]], dtype=torch.float32)
        labels = torch.tensor([obj["category_id"] for obj in annotation["annotations"]], dtype=torch.int64)

        if self.transforms:
            img = self.transforms(img)

        target = {"boxes": boxes, "labels": labels}

        return img, target

    def __len__(self):
        return len(self.imgs)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import cycle
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.applications import VGG16, ResNet50, EfficientNetB0
from tensorflow.keras.layers import GlobalAveragePooling2D
from sklearn.utils.class_weight import compute_class_weight
import pickle

# Define paths
train_dir = '/content/Thoth/Dataset/train/images'
val_dir = '/content/Thoth/Dataset/val/images'
test_dir = '/content/Thoth/Dataset/test/images'

# Create ImageDataGenerator instances for each set
datagen = ImageDataGenerator(rescale=1./255)

# Function to load data
def load_data(directory):
    generator = datagen.flow_from_directory(
        directory,
        target_size=(50, 75),
        batch_size=32,
        class_mode='categorical',
        shuffle=False
    )
    return generator

train_generator = load_data(train_dir)
val_generator = load_data(val_dir)
test_generator = load_data(test_dir)

# Extract data and labels
def extract_data(generator):
    data = []
    labels = []
    for i in range(len(generator)):
        x, y = generator.next()
        data.append(x)
        labels.append(y)
    return np.vstack(data), np.vstack(labels)

X_train, y_train = extract_data(train_generator)
X_val, y_val = extract_data(val_generator)
X_test, y_test = extract_data(test_generator)

# Calculate class weights
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(np.argmax(y_train, axis=1)),
    y=np.argmax(y_train, axis=1)
)
class_weights = dict(enumerate(class_weights))

# Define model building functions
def build_cnn_model(input_shape, num_classes):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def build_vgg_model(input_shape, num_classes):
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)
    model = Sequential([
        base_model,
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    for layer in base_model.layers:
        layer.trainable = False
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def build_resnet_model(input_shape, num_classes):
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    for layer in base_model.layers:
        layer.trainable = False
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def build_efficientnet_model(input_shape, num_classes):
    base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=input_shape)
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    for layer in base_model.layers:
        layer.trainable = False
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Dictionary to store model building functions
model_builders = {
    'CNN': build_cnn_model,
    'VGG': build_vgg_model,
    'ResNet': build_resnet_model,
    'EfficientNet': build_efficientnet_model
}

# Function to train a model
def train_model(build_model_fn, X_train, y_train, X_val, y_val, class_weights, epochs=3):
    model = build_model_fn(input_shape=(50, 75, 3), num_classes=y_train.shape[1])
    model.fit(
        X_train, y_train,
        epochs=epochs,
        validation_data=(X_val, y_val),
        class_weight=class_weights
    )
    return model

# Train and save predictions
for model_name, build_model_fn in model_builders.items():
    print(f"Training {model_name}...")
    trained_model = train_model(build_model_fn, X_train, y_train, X_val, y_val, class_weights)
    y_val_pred_probs = trained_model.predict(X_val)

    # Save true labels, predicted probabilities, and class indices
    with open(f"{model_name}_evaluation_data.pkl", "wb") as f:
        pickle.dump((y_val, y_val_pred_probs, train_generator.class_indices), f)
    print(f"Saved evaluation data for {model_name}")
```

#### 11_a) Data Loading and Preprocessing

```python
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.utils.class_weight import compute_class_weight
import pickle

# Define paths to data directories
train_dir = '/content/Thoth/Dataset/train/images'
val_dir = '/content/Thoth/Dataset/val/images'
test_dir = '/content/Thoth/Dataset/test/images'

# Create ImageDataGenerator instances for each set
datagen = ImageDataGenerator(rescale=1./255)

# Function to load data
def load_data(directory):
    generator = datagen.flow_from_directory(
        directory,
        target_size=(50, 75),
        batch_size=32,
        class_mode='categorical',
        shuffle=False
    )
    return generator

train_generator = load_data(train_dir)
val_generator = load_data(val_dir)
test_generator = load_data(test_dir)

# Extract data and labels
def extract_data(generator):
    data = []
    labels = []
    for i in range(len(generator)):
        x, y = generator.next()
        data.append(x)
        labels.append(y)
    return np.vstack(data), np.vstack(labels)

X_train, y_train = extract_data(train_generator)
X_val, y_val = extract_data(val_generator)
X_test, y_test = extract_data(test_generator)

# Calculate class weights
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(np.argmax(y_train, axis=1)),
    y=np.argmax(y_train, axis=1)
)
class_weights = dict(enumerate(class_weights))

# Save the prepared data and class weights for later use
with open('prepared_data.pkl', 'wb') as f:
    pickle.dump((X_train, y_train, X_val, y_val, X_test, y_test, class_weights, train_generator.class_indices), f)
```

#### 11_b) Model Definition and Training Function

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.applications import VGG16

# Load the prepared data and class weights
with open('prepared_data.pkl', 'rb') as f:
    X_train, y_train, X_val, y_val, X_test, y_test, class_weights, class_indices = pickle.load(f)

# Define the VGG model
def build_vgg_model(input_shape, num_classes):
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)
    model = Sequential([
        base_model,
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    for layer in base_model.layers:
        layer.trainable = False
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Build and compile the VGG model
vgg_model = build_vgg_model(input_shape=(50, 75, 3), num_classes=y_train.shape[1])

# Function to evaluate and save performance metrics
def evaluate_and_save_metrics(model, X_val, y_val, class_indices, epoch):
    y_val_pred_probs = model.predict(X_val)

    # Save true labels, predicted probabilities, and class indices
    with open(f"VGG_evaluation_data_epoch_{epoch}.pkl", "wb") as f:
        pickle.dump((y_val, y_val_pred_probs, class_indices), f)

# Train the VGG model and evaluate at each epoch
epochs = 50
for epoch in range(1, epochs + 1):
    vgg_model.fit(
        X_train, y_train,
        epochs=1,
        validation_data=(X_val, y_val),
        class_weight=class_weights,
        verbose=1
    )
    evaluate_and_save_metrics(vgg_model, X_val, y_val, class_indices, epoch)
    # Save the model after each epoch
    vgg_model.save(f"VGG_model_epoch_{epoch}.h5")
```

#### Full Training Function
1. Check what device is used for training
2. Define a set of hyperparameters to try
3. Define the number of K-Folds for Cross Validation
4. Create dataloaders from the tokenized datasets for training
5. Define the Optimizer functions
6. Define the training function
7. Run the cross validation for 4 epoch
8. Save the best hyperparameters
9. Plot the results




```python
from tqdm import tqdm
import torch
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
from transformers import T5ForConditionalGeneration, AdamW, get_linear_schedule_with_warmup
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold

# Ensure the device is set to GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define hyperparameters to try
hyperparameters = [
    {'lr': 5e-5, 'batch_size': 16, 'weight_decay': 0.01, 'warmup_steps': 0, 'optimizer': 'AdamW'},
    {'lr': 1e-4, 'batch_size': 32, 'weight_decay': 0.01, 'warmup_steps': 100, 'optimizer': 'AdamW'},
    {'lr': 3e-5, 'batch_size': 16, 'weight_decay': 0.01, 'warmup_steps': 50, 'optimizer': 'SGD'},
    {'lr': 3e-5, 'batch_size': 32, 'weight_decay': 0.01, 'warmup_steps': 100, 'optimizer': 'SGD'},
]

# K-Fold Cross Validation setup
kf = KFold(n_splits=5)

# Create Data Loaders
class TranslationDataset(Dataset):
    def __init__(self, input_ids, attention_mask, labels):
        self.input_ids = input_ids
        self.attention_mask = attention_mask
        self.labels = labels

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return {
            'input_ids': self.input_ids[idx],
            'attention_mask': self.attention_mask[idx],
            'labels': self.labels[idx]
        }

def create_data_loaders(train_idx, val_idx, hyperparameters):
    train_dataset = TranslationDataset(
        egy_train_tokens['input_ids'][train_idx],
        egy_train_tokens['attention_mask'][train_idx],
        eng_train_tokens['input_ids'][train_idx]
    )
    val_dataset = TranslationDataset(
        egy_train_tokens['input_ids'][val_idx],
        egy_train_tokens['attention_mask'][val_idx],
        eng_train_tokens['input_ids'][val_idx]
    )

    train_loader = DataLoader(train_dataset, batch_size=hyperparameters['batch_size'], shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=hyperparameters['batch_size'])

    return train_loader, val_loader

# Optimizer function
def get_optimizer(optimizer_name, model_parameters, lr, weight_decay):
    if optimizer_name == 'AdamW':
        return AdamW(model_parameters, lr=lr, weight_decay=weight_decay)
    elif optimizer_name == 'SGD':
        return optim.SGD(model_parameters, lr=lr, weight_decay=weight_decay)
    else:
        raise ValueError(f"Unsupported optimizer: {optimizer_name}")

# Training function
def train_model(hyperparameters, train_loader, val_loader):
    model = T5ForConditionalGeneration.from_pretrained('t5-small')
    model.to(device)

    optimizer = get_optimizer(hyperparameters['optimizer'], model.parameters(), hyperparameters['lr'], hyperparameters['weight_decay'])
    total_steps = len(train_loader) * num_epochs
    scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=hyperparameters['warmup_steps'], num_training_steps=total_steps)

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        for batch in tqdm(train_loader, desc=f"Training Epoch {epoch+1}/{num_epochs}"):
            optimizer.zero_grad()

            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            train_loss += loss.item()

            loss.backward()
            optimizer.step()
            scheduler.step()

        avg_train_loss = train_loss / len(train_loader)

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for batch in tqdm(val_loader, desc=f"Validating Epoch {epoch+1}/{num_epochs}"):
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)

                outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
                loss = outputs.loss
                val_loss += loss.item()

        avg_val_loss = val_loss / len(val_loader)

        print(f"Epoch {epoch+1}/{num_epochs}")
        print(f"Train Loss: {avg_train_loss:.4f}, Validation Loss: {avg_val_loss:.4f}")

    return avg_val_loss

# Run Cross-Validation with Hyperparameters
results = []
num_epochs = 4  # Reduced for quicker runs

for hyperparam in hyperparameters:
    fold_results = []
    for train_idx, val_idx in kf.split(egy_train.index):
        train_loader, val_loader = create_data_loaders(train_idx, val_idx, hyperparam)
        val_loss = train_model(hyperparam, train_loader, val_loader)
        fold_results.append(val_loss)
    avg_val_loss = sum(fold_results) / len(fold_results)
    results.append((hyperparam, avg_val_loss))
    print(f"Hyperparameters: {hyperparam}, Average Validation Loss: {avg_val_loss:.4f}")

# Determine the best hyperparameters
best_hyperparameters = min(results, key=lambda x: x[1])
print(f"Best Hyperparameters: {best_hyperparameters[0]}, with Validation Loss: {best_hyperparameters[1]:.4f}")

# Plot the results
lr_list = [res[0]['lr'] for res in results]
batch_size_list = [res[0]['batch_size'] for res in results]
optimizer_list = [res[0]['optimizer'] for res in results]
val_losses = [res[1] for res in results]

# Create a scatter plot
plt.figure(figsize=(12, 6))
for i in range(len(results)):
    plt.scatter(lr_list[i], val_losses[i], label=f"Batch Size: {batch_size_list[i]}, Optimizer: {optimizer_list[i]}", s=100)

plt.xlabel('Learning Rate')
plt.ylabel('Average Validation Loss')
plt.title('Hyperparameter Tuning Results')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch
from torch.utils.data import DataLoader, Dataset

class HieroglyphDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels['input_ids'][idx])
        return item

    def __len__(self):
        return len(self.encodings['input_ids'])

train_dataset = HieroglyphDataset(train_encodings, train_labels)
val_dataset = HieroglyphDataset(val_encodings, val_labels)
test_dataset = HieroglyphDataset(test_encodings, test_labels)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
```

```python
from transformers import AdamW
from tqdm import tqdm
import numpy as np

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)

optimizer = AdamW(model.parameters(), lr=5e-5)

# Training function with early stopping
def train_model(model, train_loader, val_loader, epochs=25, patience=3):
    model.train()
    # Recording train, and val losses for graphing learning curve
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    patience_counter = 0

    for epoch in range(epochs):
        print(f"Epoch {epoch+1}")
        train_loss = 0
        for batch in tqdm(train_loader):
            optimizer.zero_grad()
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)
            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            train_loss += loss.item()
            loss.backward()
            optimizer.step()
        train_losses.append(train_loss / len(train_loader))
        print(f"Train Loss: {train_losses[-1]}")

        # Validation
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)
                outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
                loss = outputs.loss
                val_loss += loss.item()
        val_losses.append(val_loss / len(val_loader))
        print(f"Validation Loss: {val_losses[-1]}")
        model.train()

        # Early stopping
        if val_losses[-1] < best_val_loss:
            best_val_loss = val_losses[-1]
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print("Early stopping triggered")
                break

    return train_losses, val_losses
```

###### Understanding BLEU Score

**BLEU** (Bilingual Evaluation Understudy) is a metric used to evaluate the quality of text which has been machine-translated from one language to another. It measures how many words overlap between the machine-generated translation and a set of reference translations, considering the precision of n-grams.

###### Key Aspects of BLEU:

1. **N-gram Precision**: BLEU calculates precision for different lengths of n-grams (e.g., unigrams, bigrams, trigrams).
2. **Brevity Penalty**: To prevent generating very short translations that would get high precision scores, BLEU includes a brevity penalty for translations that are shorter than the reference.
3. **Smoothing**: In cases where higher-order n-grams do not match, BLEU uses smoothing techniques to adjust scores.

###### BLEU Score Calculation

- **Unigram Precision**: Measures how many individual words match between the generated translation and the reference.
- **Bigram Precision**: Measures how many two-word sequences match.
- **Trigram Precision**, etc.

BLEU score is calculated as follows:

\[ \text{BLEU} = BP \cdot \exp \left( \sum_{n=1}^{N} w_n \log p_n \right) \]

Where:
- \( BP \) is the brevity penalty.
- \( w_n \) is the weight for the n-gram precision (commonly uniform, e.g., \( \frac{1}{N} \) for N=4).
- \( p_n \) is the precision for the n-gram length \( n \).

###### What is an Ideal BLEU Score?

- **0 to 1 Range**: BLEU scores range from 0 to 1, where 1 indicates a perfect match between the machine translation and the reference translations.
- **Interpretation**:
  - **0.0 to 0.1**: Very poor translation quality.
  - **0.1 to 0.3**: Low quality, significant errors in translation.
  - **0.3 to 0.5**: Decent translation quality, but with noticeable errors.
  - **0.5 to 0.7**: Good translation quality, with minor errors.
  - **0.7 to 0.9**: Very high-quality translation, almost no errors.
  - **0.9 to 1.0**: Near-perfect translation.

###### Example:

- **High BLEU Score**: "The cat is on the mat." vs. "The cat is on the mat."
- **Low BLEU Score**: "The cat is on the mat." vs. "The dog sits under the table."

###### Limitations of BLEU

1. **Reference Dependence**: Relies on having high-quality reference translations.
2. **Lack of Semantic Understanding**: Measures exact matches, not semantic similarity.
3. **Not Suitable for Single Sentence Evaluations**: Designed for corpus-level evaluation, not single sentences.

###### Improving BLEU Score:

1. **Quality and Quantity of Data**: More and better training data can significantly improve model performance.
2. **Model Architecture**: Using more sophisticated models can lead to better translations.
3. **Hyperparameter Tuning**: Optimizing hyperparameters like learning rate, batch size, and dropout rates.
4. **Data Augmentation**: Adding more diverse training examples.

###### Conclusion

BLEU score is a useful metric for evaluating machine translation quality, but it should be used alongside other evaluation metrics and qualitative analysis to get a comprehensive understanding of translation quality. An ideal BLEU score depends on the context and requirements of the specific task, but generally, a higher BLEU score indicates better translation quality.

#### Step 3: Attention Heatmap for the Tokenizer


```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def get_attention_scores(model, tokenizer, text, max_new_tokens=50):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt")

    # Forward pass to get outputs
    outputs = model.generate(
        input_ids=inputs.input_ids,
        attention_mask=inputs.attention_mask,
        output_attentions=True,
        return_dict_in_generate=True,
        max_new_tokens=max_new_tokens
    )

    # Get the attention scores from the output
    attentions = outputs.decoder_attentions

    return inputs, attentions

# Example text to generate attention scores using transliteration with or without prompt
text = "Translate hieroglyph transliteration to English: pr-anx nTr nfr"  # Adjust as needed
inputs, attentions = get_attention_scores(model, tokenizer, text)

def plot_attention_heatmap(inputs, attentions, tokenizer):
    # Convert attention scores to numpy array
    attention_scores = attentions[-1][0][0].detach().cpu().numpy()  # Extracting the appropriate layer and head

    # Get the tokenized input text
    tokens = tokenizer.convert_ids_to_tokens(inputs.input_ids[0])

    # Ensure the attention scores are 2D
    attention_scores = np.squeeze(attention_scores)

    # Adjust the length of tokens if necessary
    tokens = tokens[:attention_scores.shape[0]]

    # Create a heat map for the last layer's attention scores
    plt.figure(figsize=(12, 10))
    sns.heatmap(attention_scores, xticklabels=tokens, yticklabels=tokens, cmap='viridis')
    plt.title('Attention Heatmap')
    plt.xlabel('Query')
    plt.ylabel('Key')
    plt.show()

# Plot the attention heatmap
plot_attention_heatmap(inputs, attentions, tokenizer)

```

##### Analysis of heat map
The attention heatmap visualizes how the model attends to each token in the input sequence when generating each token in the output sequence. Here's an analysis of the heatmap:

   - For example, the token "glyph" in the query has high attention to the token "glyph" in the key, indicating the model's strong focus on this specific token when generating its corresponding output.

2. **Off-Diagonal Attention**:
   - Some off-diagonal cells also have high attention scores, indicating that the model is attending to other tokens beyond their direct counterparts. This could reflect the model's attempt to understand the context better.

3. **Sparse Attention**:

###### Insights from the Heatmap

- **Context Understanding**: The off-diagonal attention indicates that the model is attempting to understand the context, but there is room for improvement in how it balances attention between direct counterparts and broader contextual tokens.

```python
import pdfplumber
import re
import pandas as pd
import logging

# Set up logging
logging.basicConfig(level=logging.ERROR, format='%(asctime)s - %(levelname)s - %(message)s')

# Define the regex pattern
pattern = re.compile(
    r"\[(?P<transliteration>[^\]]+)\]"  # Match transliteration
    r"\s*(\((?P<part_of_speech>[^\)]+)\))?"  # Optionally match part of speech
    r"\s*(?P<definition>[^\{]+)"  # Match English definition
    r"\s*\{(?P<gardiner_codes>[^\}]+)\}"  # Match Gardiner codes
)

# Function to extract text from PDF using pdfplumber with enhanced error handling
def extract_text_from_pdf(pdf_path):
    text = ''
    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page_num, page in enumerate(pdf.pages):
                try:
                    page_text = page.extract_text()
                    if page_text:
                        text += page_text + '\n'  # Add newline for better separation between pages
                    else:
                        logging.warning(f"No text found on page {page_num}")
                except Exception as e:
                    logging.error(f"Error extracting text from page {page_num}: {e}")
    except Exception as e:
        logging.error(f"Error opening PDF file: {e}")
    return text

# Extract text from the provided PDF file
pdf_path = '/content/Dictionary_of_Middle_Egyptian_in_Gardine.pdf'
pdf_text = extract_text_from_pdf(pdf_path)

# Check if text extraction was successful
if pdf_text:
    # Find all matches in the text
    matches = pattern.findall(pdf_text)

    # Convert matches to a DataFrame
    entries = []
    for match in matches:
        entries.append({
            "Transliteration": match[0].strip(),
            "Part of Speech": match[1].strip() if match[1] else None,
            "Definition": match[2].strip(),
            "Gardiner Codes": match[3].strip()
        })

    df = pd.DataFrame(entries)

    # Save to CSV
    df.to_csv('parsed_dictionary_entries.csv', index=False)
    print("Parsed dictionary entries saved to 'parsed_dictionary_entries.csv'")
else:
    print("Failed to extract text from the PDF.")


```
